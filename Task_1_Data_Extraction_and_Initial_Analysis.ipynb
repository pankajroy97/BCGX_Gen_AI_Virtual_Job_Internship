```python
import pandas as pd
```

```python
final_report=pd.read_csv("/content/Final_report.csv")
final_report.head(10)
```

```python
final_report.info()
```

**Calculating Year-by-Year growth rates for Total Revenue and Net Income**

```python
final_report['Revenue_growth(%)']=final_report.groupby('Company')['Total Revenue'].pct_change()*100
final_report['Net_Income_growth(%)']=final_report.groupby('Company')['Net Income'].pct_change()*100
```

```python
final_report.head(10)
```

**mputing Null/NaN values that result from pct_change calculations with 0**

```python
final_report.fillna(0,inplace=True)
```

```python
final_report
```

**Calculating Year-by-Year growth rates for Total Assets, Total Liabilities and Cash flow from Operations Activities**

```python
final_report['Assets_growth(%)']=final_report.groupby('Company')['Total Assets'].pct_change()*100
final_report['Liabilities_growth(%)']=final_report.groupby('Company')['Total Liabilities'].pct_change()*100
final_report['cash_flow_growth(%)']=final_report.groupby('Company')['Cash Flow from Operating Activities'].pct_change()*100
```

```python
final_report
```

```python
final_report.fillna(0,inplace=True)
```
